```python
#!/usr/bin/env python
# -*- coding: utf-8 -*-

"""
figure8_rc_ukf_demo.py

Demonstration of a figure-eight (Lissajous) trajectory being learned by a 
Reservoir Computer (RC) and estimated via an Unscented Kalman Filter (UKF).

All code is self-contained so you can run it directly.
"""

import numpy as np
import matplotlib.pyplot as plt

###############################################################################
# 1) Generate a figure-eight (Lissajous) trajectory
###############################################################################
def generate_figure8_trajectory(num_steps=1000, dt=0.01, A=1.0, B=1.0, w=1.0):
    """
    Generate a parametric figure-8 (Lissajous) trajectory in 2D.

    x(t) = A * sin(w * t)
    y(t) = B * sin(2 * w * t)

    Args:
        num_steps (int): Number of discrete time steps to generate.
        dt (float): Time step between consecutive samples.
        A (float): Amplitude for x direction.
        B (float): Amplitude for y direction.
        w (float): Base frequency for the sine.

    Returns:
        ts (np.ndarray): Time array of shape (num_steps,).
        x (np.ndarray): X-values of shape (num_steps,).
        y (np.ndarray): Y-values of shape (num_steps,).
    """
    ts = np.arange(num_steps) * dt
    x = A * np.sin(w * ts)
    y = B * np.sin(2.0 * w * ts)
    return ts, x, y

###############################################################################
# 2) Define a minimal ReservoirComputer (Echo State Network) class
###############################################################################
class ReservoirComputer:
    """
    Simple implementation of a Reservoir Computer (a.k.a. Echo State Network).
    Predicts next state from the current state via a readout layer trained with ridge regression.
    """

    def __init__(
        self,
        n_inputs,
        n_reservoir=200,
        spectral_radius=0.9,
        sparsity=0.2,
        reg=1e-5,
        leak_rate=1.0,
        washout=100,
        use_bias=True,
        input_scale=0.1,
        noise_std=0.0,
        random_seed=None
    ):
        """
        Args:
            n_inputs (int): Dimensionality of input (e.g., 2 for [x, y]).
            n_reservoir (int): Number of reservoir neurons.
            spectral_radius (float): Scale the largest eigenvalue of W to this radius.
            sparsity (float): Fraction of non-zero entries in W.
            reg (float): Ridge regression factor (Tikhonov).
            leak_rate (float): Leaking rate for the reservoir update.
            washout (int): Number of initial steps to discard from training.
            use_bias (bool): Whether to add a bias term to the readout.
            input_scale (float): Scaling factor for input weight matrix W_in.
            noise_std (float): Gaussian noise on the input (optional).
            random_seed (int or None): For reproducible random initialization.
        """
        if random_seed is not None:
            np.random.seed(random_seed)

        self.n_inputs = n_inputs
        self.n_reservoir = n_reservoir
        self.spectral_radius = spectral_radius
        self.sparsity = sparsity
        self.reg = reg
        self.leak_rate = leak_rate
        self.washout = washout
        self.use_bias = use_bias
        self.input_scale = input_scale
        self.noise_std = noise_std

        # Initialize random weight matrices
        self.W_in = self._initialize_input_weights()
        self.W = self._initialize_reservoir()

        self.W_out = None  # Will be trained later

    def _initialize_input_weights(self):
        """Initialize input-to-reservoir weights in [-1, 1] scaled by input_scale."""
        return self.input_scale * (2.0 * np.random.rand(self.n_reservoir, self.n_inputs) - 1.0)

    def _initialize_reservoir(self):
        """Initialize reservoir recurrent matrix W with given sparsity and spectral radius."""
        W = np.random.rand(self.n_reservoir, self.n_reservoir) - 0.5
        # Apply sparsity
        mask = np.random.rand(self.n_reservoir, self.n_reservoir)
        W[mask > self.sparsity] = 0.0
        # Scale to spectral radius
        eigvals = np.linalg.eigvals(W)
        max_eig = np.max(np.abs(eigvals))
        if max_eig != 0.0:
            W *= (self.spectral_radius / max_eig)
        return W

    def add_noise(self, x):
        """Optionally add Gaussian noise to input vector."""
        if self.noise_std > 0:
            return x + np.random.normal(0, self.noise_std, x.shape)
        return x

    def _update_state(self, x_prev, u):
        """
        Leaky update:
        x_next = (1 - leak_rate)*x_prev + leak_rate*tanh(W_in*u + W*x_prev)
        """
        pre_activation = self.W_in.dot(u) + self.W.dot(x_prev)
        return (1 - self.leak_rate)*x_prev + self.leak_rate*np.tanh(pre_activation)

    def run_reservoir(self, input_sequence):
        """
        Run the reservoir on the entire input sequence, returning all states post-washout.

        Args:
            input_sequence (ndarray): shape (T, n_inputs)

        Returns:
            states (ndarray): shape (T - washout, n_reservoir)
        """
        T = input_sequence.shape[0]
        x = np.zeros((self.n_reservoir, 1))
        all_states = np.zeros((T, self.n_reservoir))

        for t in range(T):
            u = self.add_noise(input_sequence[t, :]).reshape(-1, 1)
            x = self._update_state(x, u)
            all_states[t] = x.ravel()

        if self.washout >= T:
            raise ValueError("washout is greater than or equal to the total timesteps.")

        return all_states[self.washout:]

    def train_readout(self, states, outputs):
        """
        Train the readout weights with ridge regression:
        W_out = (Y^T X_aug)(X_aug^T X_aug + reg*I)^{-1}

        Args:
            states (ndarray): shape (N, n_reservoir)
            outputs (ndarray): shape (N, n_outputs)
        """
        if self.use_bias:
            # Add a column of ones
            X_aug = np.hstack([states, np.ones((states.shape[0], 1))])
        else:
            X_aug = states

        # Ridge regression
        ridge_mat = self.reg * np.eye(X_aug.shape[1])
        self.W_out = (outputs.T @ X_aug) @ np.linalg.inv(X_aug.T @ X_aug + ridge_mat)

    def predict(self, states):
        """
        Given reservoir states, compute the output using W_out.

        Args:
            states (ndarray): shape (N, n_reservoir)

        Returns:
            predictions (ndarray): shape (N, n_outputs)
        """
        if self.W_out is None:
            raise ValueError("Readout weights not trained.")

        if self.use_bias:
            X_aug = np.hstack([states, np.ones((states.shape[0], 1))])
        else:
            X_aug = states

        return X_aug @ self.W_out.T

###############################################################################
# 3) Define a minimal Unscented Kalman Filter (UKF)
###############################################################################
class UnscentedKalmanFilter:
    """
    A straightforward Unscented Kalman Filter for state estimation in dimension n_dim.
    """

    def __init__(self, n_dim, Q, R, alpha=1e-3, beta=2.0, kappa=0.0):
        """
        Args:
            n_dim (int): Dimension of the state.
            Q (ndarray): Process noise covariance (n_dim x n_dim).
            R (ndarray): Measurement noise covariance (m x m).
            alpha, beta, kappa: UKF hyperparameters.
        """
        self.n_dim = n_dim
        self.Q = Q
        self.R = R

        self.alpha = alpha
        self.beta = beta
        self.kappa = kappa

        self.lambda_ = alpha**2 * (n_dim + kappa) - n_dim

        # Initialize weights
        num_sigma = 2*n_dim + 1
        self.W_m = np.full(num_sigma, 1.0/(2*(n_dim+self.lambda_)))
        self.W_c = np.full(num_sigma, 1.0/(2*(n_dim+self.lambda_)))
        self.W_m[0] = self.lambda_/(n_dim+self.lambda_)
        self.W_c[0] = self.lambda_/(n_dim+self.lambda_) + (1 - alpha**2 + beta)

        # State estimate, covariance
        self.x = np.zeros(n_dim)           # mean estimate
        self.P = np.eye(n_dim)             # covariance estimate

    def generate_sigma_points(self):
        """Generate sigma points around current state x, covariance P."""
        n = self.n_dim
        sigma_points = np.zeros((2*n + 1, n))
        sigma_points[0] = self.x

        # sqrt of (n_dim + lambda)*P
        U = np.linalg.cholesky((self.n_dim + self.lambda_)*self.P)
        for i in range(n):
            sigma_points[i+1] = self.x + U[i]
            sigma_points[n+i+1] = self.x - U[i]

        return sigma_points

    def predict(self, process_model):
        """Predict step: propagate sigma points through the process model."""
        sigma_points = self.generate_sigma_points()
        # Propagate
        sigma_points_pred = np.array([process_model(sp) for sp in sigma_points])

        # Weighted mean
        self.x = np.sum(self.W_m[:, None] * sigma_points_pred, axis=0)

        # Weighted covariance
        self.P = self.Q.copy()
        for i in range(2*self.n_dim + 1):
            diff = sigma_points_pred[i] - self.x
            self.P += self.W_c[i] * np.outer(diff, diff)

        return sigma_points_pred

    def update(self, sigma_points_pred, z, measurement_model):
        """Update step: incorporate measurement z."""
        # map sigma points to measurement space
        Zsigma = np.array([measurement_model(sp) for sp in sigma_points_pred])
        z_mean = np.sum(self.W_m[:, None] * Zsigma, axis=0)

        # measurement covariance
        P_zz = self.R.copy()
        for i in range(2*self.n_dim + 1):
            diff_z = Zsigma[i] - z_mean
            P_zz += self.W_c[i] * np.outer(diff_z, diff_z)

        # cross covariance
        P_xz = np.zeros((self.n_dim, len(z_mean)))
        for i in range(2*self.n_dim + 1):
            diff_x = sigma_points_pred[i] - self.x
            diff_z = Zsigma[i] - z_mean
            P_xz += self.W_c[i] * np.outer(diff_x, diff_z)

        # Kalman gain
        K = P_xz.dot(np.linalg.inv(P_zz))

        # Update state
        self.x += K.dot(z - z_mean)
        # Update covariance
        self.P -= K.dot(P_zz).dot(K.T)

    def get_state(self):
        """Return current state estimate."""
        return self.x

###############################################################################
# 4) Define an RC+UKF wrapper that merges the above
###############################################################################
class RC_UKF:
    """
    Hybrid that uses a trained ReservoirComputer as the process model 
    inside an Unscented Kalman Filter.
    """

    def __init__(self,
                 n_inputs,
                 n_reservoir,
                 process_noise,
                 measurement_noise,
                 rc_params=None,
                 ukf_params=None):
        if rc_params is None:
            rc_params = {}
        if ukf_params is None:
            ukf_params = {}

        # Build the RC
        self.rc = ReservoirComputer(
            n_inputs=n_inputs,
            n_reservoir=n_reservoir,
            **rc_params
        )

        # Build the UKF
        self.ukf = UnscentedKalmanFilter(
            n_dim=n_inputs,
            Q=process_noise,
            R=measurement_noise,
            **ukf_params
        )

        # Internal reservoir state
        self.reservoir_state = np.zeros((n_reservoir, 1))

        # For optional measurement noise in the measurement model
        self.measurement_model_noise_std = 0.0

    def train_reservoir(self, train_inputs, train_outputs):
        """
        Train the reservoir to map train_inputs -> train_outputs in a next-step sense.
        """
        states = self.rc.run_reservoir(train_inputs)
        valid_outputs = train_outputs[self.rc.washout:]
        self.rc.train_readout(states, valid_outputs)

    def reset_reservoir_state(self):
        """Reset the reservoir's hidden state."""
        self.reservoir_state = np.zeros((self.rc.n_reservoir, 1))

    def process_model(self, x):
        """
        The RC-based process model. 
        x is shape (n_inputs,); we feed it to the reservoir, 
        then read out the next state.
        """
        x_col = x.reshape(-1,1)
        self.reservoir_state = self.rc._update_state(self.reservoir_state, x_col)
        # next state = W_out * reservoir_state
        W_out = self.rc.W_out
        n_in = self.rc.n_inputs
        n_r = self.rc.n_reservoir

        # x_next shape (n_in,)
        if self.rc.use_bias:
            # x_next = W_out[:,:n_r]*state + W_out[:, -1]
            x_next = W_out[:, :n_r].dot(self.reservoir_state).ravel() \
                     + W_out[:, -1]
        else:
            x_next = W_out[:, :n_r].dot(self.reservoir_state).ravel()

        return x_next

    def measurement_model(self, x):
        """Full measurement of x plus optional additive noise."""
        return x + np.random.normal(0.0, self.measurement_model_noise_std, x.shape)

    def filter_step(self, measurement):
        """Perform one UKF predict+update cycle."""
        sigma_pred = self.ukf.predict(process_model=self.process_model)
        self.ukf.update(sigma_pred, measurement, measurement_model=self.measurement_model)
        return self.ukf.get_state()

###############################################################################
# 5) A small RMSE utility
###############################################################################
def compute_rmse(true_vals, est_vals):
    return np.sqrt(np.mean((true_vals - est_vals)**2))

###############################################################################
# 6) Main demonstration
###############################################################################
def main():
    # -------------------------------
    # Generate a figure-8 trajectory
    # -------------------------------
    num_steps = 1000
    dt = 0.01
    ts, x_traj, y_traj = generate_figure8_trajectory(num_steps=num_steps, dt=dt)

    # Form input-output pairs for next-step learning: (x(t), y(t)) -> (x(t+1), y(t+1))
    inputs = np.column_stack([x_traj[:-1], y_traj[:-1]])  # shape (999, 2)
    outputs = np.column_stack([x_traj[1:], y_traj[1:]])   # shape (999, 2)

    # Train/test split (e.g., 70%/30%)
    split_idx = int(0.7 * (num_steps-1))
    train_in, train_out = inputs[:split_idx], outputs[:split_idx]
    test_in, test_out = inputs[split_idx:], outputs[split_idx:]

    # ---------------------------------------
    # Setup the RC+UKF
    # We'll track a 2D state: [x, y]
    # ---------------------------------------
    process_noise = np.eye(2)*1e-4       # Q
    measurement_noise = np.eye(2)*1e-3   # R

    framework = RC_UKF(
        n_inputs=2,
        n_reservoir=200,
        process_noise=process_noise,
        measurement_noise=measurement_noise,
        rc_params = {
            "spectral_radius": 0.9,
            "leak_rate": 1.0,
            "washout": 100,
            "reg": 1e-5,
            "random_seed": 42
        },
        ukf_params = {
            "alpha": 1.0,
            "beta": 2.0,
            "kappa": 1.0   # for n=2
        }
    )

    # Train the reservoir on the training set
    framework.train_reservoir(train_in, train_out)

    # -----------------------------------
    # Now simulate the UKF on the test set
    # We'll simulate "noisy measurements"
    # -----------------------------------
    # Initialize the UKF state with the first test input
    x0, y0 = test_in[0]
    framework.ukf.x = np.array([x0, y0]) 
    framework.ukf.P = np.eye(2) * 1.0
    framework.reset_reservoir_state()

    # Let's add some measurement noise
    meas_noise_std = 0.02
    framework.measurement_model_noise_std = 0.0  # We'll do manual measurement noise below

    estimated_states = []
    for i in range(len(test_in)):
        # Construct a measurement: test_in[i] = true state at time i, plus noise
        # (Note that test_in[i] was x(t), y(t).)
        measurement = test_in[i] + np.random.normal(0, meas_noise_std, 2)

        # UKF step
        x_est = framework.filter_step(measurement)
        estimated_states.append(x_est)

    estimated_states = np.array(estimated_states)

    # We'll compare estimated_states to the "true next states" (test_out),
    # or to the "true current states" (test_in). It depends how you want to index them.
    # For plotting a trajectory, let's compare to the ground truth "test_in" or "test_out".
    # Here let's align them with test_in. The shape is the same.

    # Compute RMSE
    rmse_val = compute_rmse(test_in, estimated_states)
    print(f"RC+UKF Figure-8 Tracking RMSE: {rmse_val:.4f}")

    # -----------------------------------
    # Plot results
    # -----------------------------------
    plt.figure(figsize=(6, 6))
    plt.plot(test_in[:,0], test_in[:,1], 'b-', label='Ground Truth (Test Segment)')
    plt.plot(estimated_states[:,0], estimated_states[:,1], 'r--', label='RC+UKF Estimate')
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.title('Figure-8 Trajectory: Ground Truth vs. RC+UKF')
    plt.axis('equal')
    plt.legend()
    plt.tight_layout()
    plt.show()

if __name__ == "__main__":
    main()
```

```python
#!/usr/bin/env python
# -*- coding: utf-8 -*-

"""
flightgoggles_car_figure8_rc_ukf.py

Demonstration of:
  1) Driving a 'car' in MIT FlightGoggles with speed/steering commands 
     to produce a figure-8-like trajectory
  2) Logging the car's true (x,y)
  3) Training a Reservoir Computer on that trajectory
  4) Using the RC inside an Unscented Kalman Filter to do state estimation
  5) Plotting the results

Prerequisites:
  - You have flightgoggles.env installed (and config YAMLs set up).
  - The flightgoggles_env function is available (like in the official demos).
  - Run:   pip install numpy matplotlib
"""

import numpy as np
import matplotlib.pyplot as plt

###############################################################################
# 1) Import FlightGoggles environment
###############################################################################
try:
    from flightgoggles.env import flightgoggles_env
except ImportError:
    raise ImportError("Could not import flightgoggles_env. "
                      "Please ensure the FlightGoggles Python API is installed.")

###############################################################################
# 2) Define ReservoirComputer, UnscentedKalmanFilter, and RC_UKF
#    (Same code you already have, condensed for brevity)
###############################################################################
class ReservoirComputer:
    def __init__(
        self,
        n_inputs,
        n_reservoir=200,
        spectral_radius=0.9,
        sparsity=0.2,
        reg=1e-5,
        leak_rate=1.0,
        washout=100,
        use_bias=True,
        input_scale=0.1,
        noise_std=0.0,
        random_seed=None
    ):
        if random_seed is not None:
            np.random.seed(random_seed)

        self.n_inputs = n_inputs
        self.n_reservoir = n_reservoir
        self.spectral_radius = spectral_radius
        self.sparsity = sparsity
        self.reg = reg
        self.leak_rate = leak_rate
        self.washout = washout
        self.use_bias = use_bias
        self.input_scale = input_scale
        self.noise_std = noise_std

        # Initialize random weights
        self.W_in = self._initialize_input_weights()
        self.W = self._initialize_reservoir()
        self.W_out = None

    def _initialize_input_weights(self):
        return self.input_scale * (2.0 * np.random.rand(self.n_reservoir, self.n_inputs) - 1.0)

    def _initialize_reservoir(self):
        W = np.random.rand(self.n_reservoir, self.n_reservoir) - 0.5
        mask = np.random.rand(self.n_reservoir, self.n_reservoir)
        W[mask > self.sparsity] = 0.0
        eigvals = np.linalg.eigvals(W)
        max_eig = np.max(np.abs(eigvals))
        if max_eig != 0.0:
            W *= (self.spectral_radius / max_eig)
        return W

    def add_noise(self, x):
        if self.noise_std > 0:
            return x + np.random.normal(0, self.noise_std, x.shape)
        return x

    def _update_state(self, x_prev, u):
        pre_activation = self.W_in.dot(u) + self.W.dot(x_prev)
        return (1 - self.leak_rate)*x_prev + self.leak_rate*np.tanh(pre_activation)

    def run_reservoir(self, input_sequence):
        T = input_sequence.shape[0]
        x = np.zeros((self.n_reservoir, 1))
        all_states = np.zeros((T, self.n_reservoir))
        for t in range(T):
            u = self.add_noise(input_sequence[t]).reshape(-1,1)
            x = self._update_state(x, u)
            all_states[t] = x.ravel()
        if self.washout >= T:
            raise ValueError("washout is >= number of timesteps.")
        return all_states[self.washout:]

    def train_readout(self, states, outputs):
        if self.use_bias:
            X_aug = np.hstack([states, np.ones((states.shape[0],1))])
        else:
            X_aug = states
        ridge_mat = self.reg * np.eye(X_aug.shape[1])
        self.W_out = (outputs.T @ X_aug) @ np.linalg.inv(X_aug.T @ X_aug + ridge_mat)

    def predict(self, states):
        if self.W_out is None:
            raise ValueError("Readout not trained.")
        if self.use_bias:
            X_aug = np.hstack([states, np.ones((states.shape[0],1))])
        else:
            X_aug = states
        return X_aug @ self.W_out.T


class UnscentedKalmanFilter:
    def __init__(self, n_dim, Q, R, alpha=1.0, beta=2.0, kappa=1.0):
        self.n_dim = n_dim
        self.Q = Q
        self.R = R
        self.alpha = alpha
        self.beta = beta
        self.kappa = kappa
        self.lambda_ = alpha**2*(n_dim+kappa) - n_dim

        num_sigma = 2*n_dim + 1
        self.W_m = np.full(num_sigma, 1.0/(2*(n_dim+self.lambda_)))
        self.W_c = np.full(num_sigma, 1.0/(2*(n_dim+self.lambda_)))
        self.W_m[0] = self.lambda_/(n_dim+self.lambda_)
        self.W_c[0] = self.lambda_/(n_dim+self.lambda_) + (1 - alpha**2 + beta)

        self.x = np.zeros(n_dim)
        self.P = np.eye(n_dim)

    def generate_sigma_points(self):
        n = self.n_dim
        sigma_points = np.zeros((2*n+1, n))
        sigma_points[0] = self.x
        # cholesky of (n_dim+lambda)*P
        U = np.linalg.cholesky((n+self.lambda_)*self.P)
        for i in range(n):
            sigma_points[i+1] = self.x + U[i]
            sigma_points[n+i+1] = self.x - U[i]
        return sigma_points

    def predict(self, process_model):
        sigma_pts = self.generate_sigma_points()
        sigma_pred = np.array([process_model(sp) for sp in sigma_pts])
        # Weighted mean
        self.x = np.sum(self.W_m[:,None]*sigma_pred, axis=0)
        # Weighted covariance
        self.P = self.Q.copy()
        for i in range(2*self.n_dim+1):
            diff = sigma_pred[i] - self.x
            self.P += self.W_c[i]*np.outer(diff,diff)
        return sigma_pred

    def update(self, sigma_pred, z, measurement_model):
        Zsigma = np.array([measurement_model(sp) for sp in sigma_pred])
        z_mean = np.sum(self.W_m[:,None]*Zsigma, axis=0)
        # measurement covariance
        P_zz = self.R.copy()
        for i in range(2*self.n_dim+1):
            diff_z = Zsigma[i] - z_mean
            P_zz += self.W_c[i]*np.outer(diff_z,diff_z)
        # cross-cov
        P_xz = np.zeros((self.n_dim, z_mean.shape[0]))
        for i in range(2*self.n_dim+1):
            diff_x = sigma_pred[i] - self.x
            diff_z = Zsigma[i] - z_mean
            P_xz += self.W_c[i]*np.outer(diff_x,diff_z)
        # Kalman gain
        K = P_xz @ np.linalg.inv(P_zz)
        self.x += K.dot(z - z_mean)
        self.P -= K.dot(P_zz).dot(K.T)

    def get_state(self):
        return self.x


class RC_UKF:
    def __init__(self,
                 n_inputs,
                 n_reservoir,
                 process_noise,
                 measurement_noise,
                 rc_params=None,
                 ukf_params=None):
        if rc_params is None:
            rc_params = {}
        if ukf_params is None:
            ukf_params = {}

        self.rc = ReservoirComputer(
            n_inputs=n_inputs,
            n_reservoir=n_reservoir,
            **rc_params
        )

        self.ukf = UnscentedKalmanFilter(
            n_dim=n_inputs,
            Q=process_noise,
            R=measurement_noise,
            **ukf_params
        )

        self.reservoir_state = np.zeros((n_reservoir,1))
        self.measurement_model_noise_std = 0.0

    def train_reservoir(self, inputs, outputs):
        states = self.rc.run_reservoir(inputs)
        valid_outputs = outputs[self.rc.washout:]
        self.rc.train_readout(states, valid_outputs)

    def reset_reservoir_state(self):
        self.reservoir_state = np.zeros((self.rc.n_reservoir,1))

    def process_model(self, x):
        x_col = x.reshape(-1,1)
        self.reservoir_state = self.rc._update_state(self.reservoir_state, x_col)
        W_out = self.rc.W_out
        if self.rc.use_bias:
            x_next = W_out[:, :self.rc.n_reservoir].dot(self.reservoir_state).ravel() \
                     + W_out[:,-1]
        else:
            x_next = W_out[:, :self.rc.n_reservoir].dot(self.reservoir_state).ravel()
        return x_next

    def measurement_model(self, x):
        return x + np.random.normal(0.0, self.measurement_model_noise_std, x.shape)

    def filter_step(self, measurement):
        sigma_pred = self.ukf.predict(self.process_model)
        self.ukf.update(sigma_pred, measurement, self.measurement_model)
        return self.ukf.get_state()

def compute_rmse(true_vals, est_vals):
    return np.sqrt(np.mean((true_vals - est_vals)**2))


###############################################################################
# 3) Main function to run a car in FlightGoggles, gather data, train RC, do UKF
###############################################################################
def main_figure8_flightgoggles():
    """
    1) Launch a FlightGoggles environment with a "car".
    2) Drive the car in a ~figure-8 shape by applying time-varying steering commands.
    3) Log (x, y) from the environment.
    4) Train an RC to model next-step dynamics.
    5) Use the RC inside a UKF for state estimation on a test portion of the data.
    6) Plot the ground truth vs. RC+UKF estimates.
    """

    # 1) Initialize environment
    print("Starting FlightGoggles environment...")
    env = flightgoggles_env(
        cfg_dir="../config",                      # Adjust if needed
        cfg_fgclient="FlightGogglesClient_demo_car.yaml",  # or your .yaml
        cfg_car="carDynamicsSim.yaml"
    )

    # 2) Drive the car in a figure-8 pattern
    logs = []
    num_steps = 2000
    dt = 0.01
    for step in range(num_steps):
        t = step * dt
        # Example steering to produce a rough figure-8:
        # speed is moderate, steering oscillates with 2 frequencies
        speed_cmd = 1.2
        steering_angle_cmd = 0.6 * np.sin(1.0*t) * np.sin(0.5*t)

        env.proceed(
            vehicle_id="car1",
            speed_command=speed_cmd,
            steering_angle_command=steering_angle_cmd,
            duration=dt
        )

        # Get ground-truth position from environment
        # Suppose env.get_state("car1") -> [x, y, z, qx, qy, qz, qw]
        pose = env.get_state("car1")
        x = pose["position"][0]
        y = pose["position"][1]
        logs.append([x,y])

    logs = np.array(logs)  # shape (num_steps, 2)

    # (Optional) You can plot or save logs directly via FlightGoggles API:
    # env.plot_state(vehicle_id="car1", attribute="position")
    # env.save_logs(vehicle_id="car1", save_dir="data/")

    print("Simulation done. Closing environment.")
    env.close()

    # 3) Train a reservoir on the logs (the "model" for the next-step dynamics)
    inputs = logs[:-1]
    outputs = logs[1:]
    N = len(inputs)
    split_idx = int(0.5*N)
    train_in, train_out = inputs[:split_idx], outputs[:split_idx]
    test_in, test_out   = inputs[split_idx:], outputs[split_idx:]

    # Setup RC+UKF
    process_noise = np.eye(2)*1e-4
    measurement_noise = np.eye(2)*1e-3

    framework = RC_UKF(
        n_inputs=2,
        n_reservoir=200,
        process_noise=process_noise,
        measurement_noise=measurement_noise,
        rc_params={
            "spectral_radius": 0.9,
            "leak_rate": 1.0,
            "washout": 100,
            "reg": 1e-5,
            "random_seed": 42
        },
        ukf_params={
            "alpha": 1.0,
            "beta": 2.0,
            "kappa": 1.0
        }
    )

    print("Training reservoir on 70% of the data...")
    framework.train_reservoir(train_in, train_out)
    print("Reservoir training complete.")

    # 4) Use the test portion for filtering
    x0, y0 = test_in[0]
    framework.ukf.x = np.array([x0, y0]) 
    framework.ukf.P = np.eye(2)*1.0
    framework.reset_reservoir_state()

    # If you want to add measurement noise in the measurement model:
    meas_noise_std = 0.02
    framework.measurement_model_noise_std = 0.0  # We'll add noise manually below

    estimated_states = []
    for i in range(len(test_in)):
        # Synthetic measurement = ground truth + noise
        measurement = test_in[i] + np.random.normal(0, meas_noise_std, 2)
        x_est = framework.filter_step(measurement)
        estimated_states.append(x_est)
    estimated_states = np.array(estimated_states)

    # 5) Evaluate and plot
    rmse_val = compute_rmse(test_in, estimated_states)
    print(f"RC+UKF Car Figure-8 Tracking RMSE: {rmse_val:.4f}")

    plt.figure(figsize=(6,6))
    plt.plot(test_in[:,0], test_in[:,1], 'b-', label='Ground Truth (Test)')
    plt.plot(estimated_states[:,0], estimated_states[:,1], 'r--', label='RC+UKF Estimate')
    plt.title("FlightGoggles Car: Figure-8 + RC+UKF")
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.axis('equal')
    plt.legend()
    plt.tight_layout()
    plt.show()


if __name__ == "__main__":
    main_figure8_flightgoggles()
```

```python
#!/usr/bin/env python
# -*- coding: utf-8 -*-

"""
flightgoggles_car_pid_figure8_rc_ukf.py

Demonstration of:
  1) Driving a 'car' in MIT FlightGoggles in a figure-8 with a minimal heading PID.
  2) Logging the car's (x,y) trajectory.
  3) Training a Reservoir Computer on that trajectory for next-step prediction.
  4) Using the trained RC as a process model inside an Unscented Kalman Filter (UKF).
  5) Plotting training data vs. test data vs. RC+UKF estimates.

Requirements:
  - `flightgoggles.env` installed + config YAMLs that define "car1" in your scene
  - `pip install numpy matplotlib`
"""

import numpy as np
import matplotlib.pyplot as plt

###############################################################################
# 1) FlightGoggles environment import
###############################################################################
try:
    from flightgoggles.env import flightgoggles_env
except ImportError:
    raise ImportError("Could not import flightgoggles_env. Make sure the FlightGoggles Python API is installed.")

###############################################################################
# 2) Helper: Convert Orientation (qx,qy,qz,qw) to heading (yaw)
###############################################################################
def quaternion_to_yaw(qx, qy, qz, qw):
    """
    Compute yaw angle in radians about Z axis from a quaternion,
    assuming the environment uses an upright coordinate (NED or NWU).
    """
    siny_cosp = 2.0 * (qw*qz + qx*qy)
    cosy_cosp = 1.0 - 2.0*(qy*qy + qz*qz)
    yaw = np.arctan2(siny_cosp, cosy_cosp)
    return yaw

###############################################################################
# 3) ReservoirComputer, UnscentedKalmanFilter, RC_UKF classes
###############################################################################
class ReservoirComputer:
    """
    A straightforward Echo State Network (ESN) to predict next state (x_{t+1})
    from current state (x_t), trained by ridge regression.
    """
    def __init__(
        self,
        n_inputs,
        n_reservoir=200,
        spectral_radius=0.9,
        sparsity=0.2,
        reg=1e-5,
        leak_rate=1.0,
        washout=100,
        use_bias=True,
        input_scale=0.1,
        noise_std=0.0,
        random_seed=None
    ):
        if random_seed is not None:
            np.random.seed(random_seed)

        self.n_inputs = n_inputs
        self.n_reservoir = n_reservoir
        self.spectral_radius = spectral_radius
        self.sparsity = sparsity
        self.reg = reg
        self.leak_rate = leak_rate
        self.washout = washout
        self.use_bias = use_bias
        self.input_scale = input_scale
        self.noise_std = noise_std

        # Initialize random weights
        self.W_in = self._initialize_input_weights()
        self.W = self._initialize_reservoir()
        self.W_out = None

    def _initialize_input_weights(self):
        return self.input_scale * (2.0*np.random.rand(self.n_reservoir, self.n_inputs) - 1.0)

    def _initialize_reservoir(self):
        W = np.random.rand(self.n_reservoir, self.n_reservoir) - 0.5
        mask = np.random.rand(self.n_reservoir, self.n_reservoir)
        W[mask > self.sparsity] = 0.0
        # Scale to spectral radius
        eigvals = np.linalg.eigvals(W)
        max_eig = np.max(np.abs(eigvals))
        if max_eig != 0.0:
            W *= (self.spectral_radius / max_eig)
        return W

    def add_noise(self, x):
        if self.noise_std > 0:
            return x + np.random.normal(0, self.noise_std, x.shape)
        return x

    def _update_state(self, x_prev, u):
        """
        Leaky update:
          x_next = (1 - leak_rate)*x_prev + leak_rate*tanh(W_in*u + W*x_prev)
        """
        pre_activation = self.W_in.dot(u) + self.W.dot(x_prev)
        return (1 - self.leak_rate)*x_prev + self.leak_rate*np.tanh(pre_activation)

    def run_reservoir(self, input_sequence):
        """
        Pass the entire input time-series through the reservoir, returning
        states after the washout period.

        input_sequence shape: (T, n_inputs)
        returns: states shape (T - washout, n_reservoir)
        """
        T = input_sequence.shape[0]
        x = np.zeros((self.n_reservoir, 1))
        all_states = np.zeros((T, self.n_reservoir))
        for t in range(T):
            u = self.add_noise(input_sequence[t]).reshape(-1,1)
            x = self._update_state(x, u)
            all_states[t] = x.ravel()
        if self.washout >= T:
            raise ValueError("washout >= total timesteps!")
        return all_states[self.washout:]

    def train_readout(self, states, outputs):
        """
        Train the readout layer via ridge regression:
          W_out = (Y^T X_aug)(X_aug^T X_aug + reg*I)^{-1}
        """
        if self.use_bias:
            X_aug = np.hstack([states, np.ones((states.shape[0],1))])
        else:
            X_aug = states
        ridge_mat = self.reg * np.eye(X_aug.shape[1])
        self.W_out = (outputs.T @ X_aug) @ np.linalg.inv(X_aug.T @ X_aug + ridge_mat)

    def predict(self, states):
        """
        Once trained, produce outputs from a batch of reservoir states.
        """
        if self.W_out is None:
            raise ValueError("Readout not trained.")
        if self.use_bias:
            X_aug = np.hstack([states, np.ones((states.shape[0],1))])
        else:
            X_aug = states
        return X_aug @ self.W_out.T


class UnscentedKalmanFilter:
    """
    A simple UKF implementation for state dimension n_dim.
    """
    def __init__(self, n_dim, Q, R, alpha=1.0, beta=2.0, kappa=1.0):
        self.n_dim = n_dim
        self.Q = Q
        self.R = R
        self.alpha = alpha
        self.beta = beta
        self.kappa = kappa
        self.lambda_ = alpha**2*(n_dim + kappa) - n_dim

        num_sigma = 2*n_dim + 1
        self.W_m = np.full(num_sigma, 1.0/(2*(n_dim+self.lambda_)))
        self.W_c = np.full(num_sigma, 1.0/(2*(n_dim+self.lambda_)))
        self.W_m[0] = self.lambda_/(n_dim+self.lambda_)
        self.W_c[0] = self.lambda_/(n_dim+self.lambda_) + (1 - alpha**2 + beta)

        self.x = np.zeros(n_dim)
        self.P = np.eye(n_dim)

    def generate_sigma_points(self):
        n = self.n_dim
        sigma_pts = np.zeros((2*n+1, n))
        sigma_pts[0] = self.x
        U = np.linalg.cholesky((n + self.lambda_)*self.P)
        for i in range(n):
            sigma_pts[i+1]    = self.x + U[i]
            sigma_pts[n+i+1] = self.x - U[i]
        return sigma_pts

    def predict(self, process_model):
        sigma_pts = self.generate_sigma_points()
        sigma_pred = np.array([process_model(sp) for sp in sigma_pts])
        # Weighted mean
        self.x = np.sum(self.W_m[:,None]*sigma_pred, axis=0)
        # Weighted covariance
        self.P = self.Q.copy()
        for i in range(2*self.n_dim+1):
            diff = sigma_pred[i] - self.x
            self.P += self.W_c[i]*np.outer(diff,diff)
        return sigma_pred

    def update(self, sigma_pred, z, measurement_model):
        Zsigma = np.array([measurement_model(sp) for sp in sigma_pred])
        z_mean = np.sum(self.W_m[:,None]*Zsigma, axis=0)

        # measurement covariance
        P_zz = self.R.copy()
        for i in range(2*self.n_dim+1):
            diff_z = Zsigma[i] - z_mean
            P_zz += self.W_c[i]*np.outer(diff_z,diff_z)

        # cross covariance
        P_xz = np.zeros((self.n_dim, z_mean.shape[0]))
        for i in range(2*self.n_dim+1):
            diff_x = sigma_pred[i] - self.x
            diff_z = Zsigma[i] - z_mean
            P_xz += self.W_c[i]*np.outer(diff_x,diff_z)

        # Kalman gain
        K = P_xz @ np.linalg.inv(P_zz)

        # update
        self.x += K.dot(z - z_mean)
        self.P -= K.dot(P_zz).dot(K.T)

    def get_state(self):
        return self.x


class RC_UKF:
    """
    Wrapper that uses a ReservoirComputer as the process_model inside a UKF.
    """
    def __init__(self,
                 n_inputs,
                 n_reservoir,
                 process_noise,
                 measurement_noise,
                 rc_params=None,
                 ukf_params=None):
        if rc_params is None:
            rc_params = {}
        if ukf_params is None:
            ukf_params = {}

        # Build the RC
        self.rc = ReservoirComputer(n_inputs=n_inputs,
                                    n_reservoir=n_reservoir,
                                    **rc_params)
        # Build the UKF
        self.ukf = UnscentedKalmanFilter(n_dim=n_inputs,
                                         Q=process_noise,
                                         R=measurement_noise,
                                         **ukf_params)
        self.reservoir_state = np.zeros((n_reservoir,1))
        self.measurement_model_noise_std = 0.0

    def train_reservoir(self, inputs, outputs):
        states = self.rc.run_reservoir(inputs)
        valid_outputs = outputs[self.rc.washout:]
        self.rc.train_readout(states, valid_outputs)

    def reset_reservoir_state(self):
        self.reservoir_state = np.zeros((self.rc.n_reservoir,1))

    def process_model(self, x):
        # feed x to ESN
        x_col = x.reshape(-1,1)
        self.reservoir_state = self.rc._update_state(self.reservoir_state, x_col)

        # read out next state
        W_out = self.rc.W_out
        if self.rc.use_bias:
            x_next = (W_out[:, :self.rc.n_reservoir].dot(self.reservoir_state).ravel()
                      + W_out[:, -1])
        else:
            x_next = W_out[:, :self.rc.n_reservoir].dot(self.reservoir_state).ravel()
        return x_next

    def measurement_model(self, x):
        return x + np.random.normal(0.0, self.measurement_model_noise_std, x.shape)

    def filter_step(self, measurement):
        sigma_pred = self.ukf.predict(self.process_model)
        self.ukf.update(sigma_pred, measurement, self.measurement_model)
        return self.ukf.get_state()

###############################################################################
# Utility function
###############################################################################
def compute_rmse(true_vals, est_vals):
    return np.sqrt(np.mean((true_vals - est_vals)**2))

###############################################################################
# 4) Main figure-8 with a heading PID, then RC+UKF
###############################################################################
def main():
    """
    1) Start FlightGoggles with a 'car'.
    2) Use a minimal heading PID to drive the car along a known figure-8 reference in (x_r, y_r).
    3) Log (x,y) from environment.
    4) Train an RC on that data.
    5) Use the RC inside a UKF to filter the test portion.
    6) Plot results.
    """

    #######################
    # Initialize environment
    #######################
    print("Starting FlightGoggles environment...")
    env = flightgoggles_env(
        cfg_dir="../config",
        cfg_fgclient="FlightGogglesClient_demo_car.yaml",
        cfg_car="carDynamicsSim.yaml"
    )

    #######################
    # Define figure-8 reference
    #######################
    def x_ref(t, R=5.0, w=0.2):
        return R*np.sin(w*t)
    def y_ref(t, R=5.0, w=0.2):
        return R*np.sin(2*w*t)
    # derivatives for heading calc
    def dx_ref(t, R=5.0, w=0.2):
        return R*w*np.cos(w*t)
    def dy_ref(t, R=5.0, w=0.2):
        return 2*R*w*np.cos(2*w*t)

    # heading P-gain
    Kp_heading = 1.5
    speed_cmd = 1.0
    logs = []

    num_steps = 3000
    dt = 0.01

    #######################
    # Step 1: Drive the car with a heading PID
    #######################
    for step in range(num_steps):
        t = step*dt
        # reference (x_r, y_r)
        xr = x_ref(t)
        yr = y_ref(t)
        # derivatives for heading
        dxr = dx_ref(t)
        dyr = dy_ref(t)

        # desired heading
        theta_r = np.arctan2(dyr, dxr)

        # get actual pose
        pose = env.get_state("car1")
        x_act = pose["position"][0]
        y_act = pose["position"][1]
        theta_act = pose["heading"]

        # heading error
        heading_err = theta_r - theta_act
        # wrap to (-pi, pi)
        heading_err = (heading_err + np.pi) % (2*np.pi) - np.pi

        # steering = Kp * heading_err
        steering_angle_cmd = Kp_heading * heading_err
        # clamp
        steering_angle_cmd = np.clip(steering_angle_cmd, -1.0, 1.0)

        # step environment
        env.proceed(
            vehicle_id="car1",
            speed_command=speed_cmd,
            steering_angle_command=steering_angle_cmd,
            duration=dt
        )

        # log
        logs.append([x_act, y_act])

    logs = np.array(logs)  # shape (num_steps,2)
    print("Simulation done. Closing environment.")
    env.close()

    #######################
    # Step 2: Train RC + UKF
    #######################
    inputs = logs[:-1]
    outputs = logs[1:]
    N = len(inputs)
    split_idx = int(0.7*N)
    train_in, train_out = inputs[:split_idx], outputs[:split_idx]
    test_in, test_out   = inputs[split_idx:], outputs[split_idx:]

    # Setup RC+UKF
    process_noise = np.eye(2)*1e-4
    measurement_noise = np.eye(2)*1e-3

    framework = RC_UKF(
        n_inputs=2,
        n_reservoir=200,
        process_noise=process_noise,
        measurement_noise=measurement_noise,
        rc_params={
            "spectral_radius": 0.9,
            "leak_rate": 1.0,
            "washout": 100,
            "reg": 1e-5,
            "random_seed": 42
        },
        ukf_params={
            "alpha": 1.0,
            "beta": 2.0,
            "kappa": 1.0
        }
    )

    print("Training reservoir on 70% of data ...")
    framework.train_reservoir(train_in, train_out)
    print("Reservoir training complete.")

    # run filter on test portion
    x0, y0 = test_in[0]
    framework.ukf.x = np.array([x0, y0])
    framework.ukf.P = np.eye(2)*1.0
    framework.reset_reservoir_state()

    # measurement noise
    meas_noise_std = 0.02
    framework.measurement_model_noise_std = 0.0  # we add noise manually

    estimated_states = []
    for i in range(len(test_in)):
        measurement = test_in[i] + np.random.normal(0, meas_noise_std, 2)
        x_est = framework.filter_step(measurement)
        estimated_states.append(x_est)
    estimated_states = np.array(estimated_states)

    rmse_val = compute_rmse(test_in, estimated_states)
    print(f"RC+UKF Car Figure-8 Tracking RMSE: {rmse_val:.4f}")

    #######################
    # Plot
    #######################
    plt.figure(figsize=(6,6))

    plt.plot(train_in[:,0], train_in[:,1], 'g-', label='Ground Truth (Train)')
    plt.plot(test_in[:,0], test_in[:,1], 'b-', label='Ground Truth (Test)')
    plt.plot(estimated_states[:,0], estimated_states[:,1], 'r--', label='RC+UKF Estimate')
    plt.title("FlightGoggles Car: Figure-8 + RC+UKF (PID-based)")
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.axis('equal')
    plt.legend()
    plt.tight_layout()
    plt.show()

if __name__ == "__main__":
    main()
```

```python
#!/usr/bin/env python
# -*- coding: utf-8 -*-

"""
flightgoggles_car_pid_figure8_rc_ukf.py

Forces a clearer figure-8 path in FlightGoggles by:
  - Lower speed, 
  - Smaller amplitude & frequency for reference,
  - Reasonable PID heading gain,
  - Enough timesteps to see a full "8" crossing.

Then logs (x,y), trains a Reservoir Computer, uses it in a UKF, and plots.

Requires:
  - flightgoggles.env installed & config YAML for a car "car1"
  - pip install numpy matplotlib
"""

import numpy as np
import matplotlib.pyplot as plt

###############################################################################
# 1) Import FlightGoggles environment
###############################################################################
try:
    from flightgoggles.env import flightgoggles_env
except ImportError:
    raise ImportError("Could not import flightgoggles_env. "
                      "Please ensure the FlightGoggles Python API is installed.")


###############################################################################
# 2) Helper: Convert heading from environment's dictionary
###############################################################################
def get_heading_from_pose(pose):
    """
    The environment returns:
      pose = {
        "timestamp": ...,
        "position": array([...]),
        "velocity": array([...]),
        "heading": float,
        "speed": float,
        "steering_angle": ...
      }
    We simply return pose["heading"] as the actual heading in radians.
    """
    return pose["heading"]


###############################################################################
# 3) Classes: ReservoirComputer, UnscentedKalmanFilter, RC_UKF, etc.
###############################################################################
class ReservoirComputer:
    def __init__(
        self,
        n_inputs,
        n_reservoir=200,
        spectral_radius=0.9,
        sparsity=0.2,
        reg=1e-5,
        leak_rate=1.0,
        washout=100,
        use_bias=True,
        input_scale=0.1,
        noise_std=0.0,
        random_seed=None
    ):
        if random_seed is not None:
            np.random.seed(random_seed)

        self.n_inputs = n_inputs
        self.n_reservoir = n_reservoir
        self.spectral_radius = spectral_radius
        self.sparsity = sparsity
        self.reg = reg
        self.leak_rate = leak_rate
        self.washout = washout
        self.use_bias = use_bias
        self.input_scale = input_scale
        self.noise_std = noise_std

        # Initialize random weights
        self.W_in = self._initialize_input_weights()
        self.W = self._initialize_reservoir()
        self.W_out = None

    def _initialize_input_weights(self):
        return self.input_scale * (2.0 * np.random.rand(self.n_reservoir, self.n_inputs) - 1.0)

    def _initialize_reservoir(self):
        W = np.random.rand(self.n_reservoir, self.n_reservoir) - 0.5
        mask = np.random.rand(self.n_reservoir, self.n_reservoir)
        W[mask > self.sparsity] = 0.0
        # spectral radius scaling
        eigvals = np.linalg.eigvals(W)
        max_eig = np.max(np.abs(eigvals))
        if max_eig != 0.0:
            W *= (self.spectral_radius / max_eig)
        return W

    def add_noise(self, x):
        if self.noise_std > 0:
            return x + np.random.normal(0, self.noise_std, x.shape)
        return x

    def _update_state(self, x_prev, u):
        pre_activation = self.W_in.dot(u) + self.W.dot(x_prev)
        return (1 - self.leak_rate)*x_prev + self.leak_rate*np.tanh(pre_activation)

    def run_reservoir(self, input_sequence):
        T = input_sequence.shape[0]
        x = np.zeros((self.n_reservoir,1))
        all_states = np.zeros((T, self.n_reservoir))
        for t in range(T):
            u = self.add_noise(input_sequence[t]).reshape(-1,1)
            x = self._update_state(x, u)
            all_states[t] = x.ravel()
        if self.washout >= T:
            raise ValueError("washout >= total timesteps.")
        return all_states[self.washout:]

    def train_readout(self, states, outputs):
        if self.use_bias:
            X_aug = np.hstack([states, np.ones((states.shape[0],1))])
        else:
            X_aug = states
        ridge_mat = self.reg * np.eye(X_aug.shape[1])
        self.W_out = (outputs.T @ X_aug) @ np.linalg.inv(X_aug.T @ X_aug + ridge_mat)

    def predict(self, states):
        if self.W_out is None:
            raise ValueError("Readout not trained.")
        if self.use_bias:
            X_aug = np.hstack([states, np.ones((states.shape[0],1))])
        else:
            X_aug = states
        return X_aug @ self.W_out.T


class UnscentedKalmanFilter:
    def __init__(self, n_dim, Q, R, alpha=1.0, beta=2.0, kappa=1.0):
        self.n_dim = n_dim
        self.Q = Q
        self.R = R
        self.alpha = alpha
        self.beta = beta
        self.kappa = kappa
        self.lambda_ = alpha**2*(n_dim + kappa) - n_dim

        num_sigma = 2*n_dim + 1
        self.W_m = np.full(num_sigma, 1.0/(2*(n_dim+self.lambda_)))
        self.W_c = np.full(num_sigma, 1.0/(2*(n_dim+self.lambda_)))
        self.W_m[0] = self.lambda_/(n_dim+self.lambda_)
        self.W_c[0] = self.lambda_/(n_dim+self.lambda_) + (1 - alpha**2 + beta)

        self.x = np.zeros(n_dim)
        self.P = np.eye(n_dim)

    def generate_sigma_points(self):
        n = self.n_dim
        sigma_pts = np.zeros((2*n+1, n))
        sigma_pts[0] = self.x
        U = np.linalg.cholesky((n + self.lambda_)*self.P)
        for i in range(n):
            sigma_pts[i+1]    = self.x + U[i]
            sigma_pts[n+i+1] = self.x - U[i]
        return sigma_pts

    def predict(self, process_model):
        sigma_pts = self.generate_sigma_points()
        sigma_pred = np.array([process_model(sp) for sp in sigma_pts])
        self.x = np.sum(self.W_m[:,None]*sigma_pred, axis=0)
        self.P = self.Q.copy()
        for i in range(2*self.n_dim+1):
            diff = sigma_pred[i] - self.x
            self.P += self.W_c[i]*np.outer(diff,diff)
        return sigma_pred

    def update(self, sigma_pred, z, measurement_model):
        Zsigma = np.array([measurement_model(sp) for sp in sigma_pred])
        z_mean = np.sum(self.W_m[:,None]*Zsigma, axis=0)

        P_zz = self.R.copy()
        for i in range(2*self.n_dim+1):
            diff_z = Zsigma[i] - z_mean
            P_zz += self.W_c[i]*np.outer(diff_z,diff_z)

        P_xz = np.zeros((self.n_dim, z_mean.shape[0]))
        for i in range(2*self.n_dim+1):
            diff_x = sigma_pred[i] - self.x
            diff_z = Zsigma[i] - z_mean
            P_xz += self.W_c[i]*np.outer(diff_x,diff_z)

        K = P_xz @ np.linalg.inv(P_zz)
        self.x += K.dot(z - z_mean)
        self.P -= K.dot(P_zz).dot(K.T)

    def get_state(self):
        return self.x


class RC_UKF:
    def __init__(self,
                 n_inputs,
                 n_reservoir,
                 process_noise,
                 measurement_noise,
                 rc_params=None,
                 ukf_params=None):
        if rc_params is None:
            rc_params = {}
        if ukf_params is None:
            ukf_params = {}

        self.rc = ReservoirComputer(
            n_inputs=n_inputs,
            n_reservoir=n_reservoir,
            **rc_params
        )
        self.ukf = UnscentedKalmanFilter(
            n_dim=n_inputs,
            Q=process_noise,
            R=measurement_noise,
            **ukf_params
        )
        self.reservoir_state = np.zeros((n_reservoir,1))
        self.measurement_model_noise_std = 0.0

    def train_reservoir(self, inputs, outputs):
        states = self.rc.run_reservoir(inputs)
        valid_outputs = outputs[self.rc.washout:]
        self.rc.train_readout(states, valid_outputs)

    def reset_reservoir_state(self):
        self.reservoir_state = np.zeros((self.rc.n_reservoir,1))

    def process_model(self, x):
        x_col = x.reshape(-1,1)
        self.reservoir_state = self.rc._update_state(self.reservoir_state, x_col)
        W_out = self.rc.W_out
        if self.rc.use_bias:
            x_next = (W_out[:, :self.rc.n_reservoir].dot(self.reservoir_state).ravel()
                      + W_out[:,-1])
        else:
            x_next = W_out[:, :self.rc.n_reservoir].dot(self.reservoir_state).ravel()
        return x_next

    def measurement_model(self, x):
        return x + np.random.normal(0.0, self.measurement_model_noise_std, x.shape)

    def filter_step(self, measurement):
        sigma_pred = self.ukf.predict(self.process_model)
        self.ukf.update(sigma_pred, measurement, self.measurement_model)
        return self.ukf.get_state()


def compute_rmse(true_vals, est_vals):
    return np.sqrt(np.mean((true_vals - est_vals)**2))


###############################################################################
# 4) Main: Force a gentler figure-8, lower speed, etc.
###############################################################################
def main():
    print("Starting FlightGoggles environment...")
    env = flightgoggles_env(
        cfg_dir="../config",  # adjust to your path
        cfg_fgclient="FlightGogglesClient_demo_car.yaml",
        cfg_car="carDynamicsSim.yaml"
    )

    # We'll define a param figure-8: x_r(t)=R sin(w t), y_r(t)=R sin(2 w t)
    # Then do a heading P-controller so the car tries to track that path.

    # Lower amplitude and frequency, so we get a smaller & slower loop:
    R = 3.0
    w = 0.1
    speed_cmd = 0.7  # slower speed => easier turning
    Kp_heading = 2.0 # increase heading gain so it tracks better

    def x_ref(t):
        return R*np.sin(w*t)
    def y_ref(t):
        return R*np.sin(2*w*t)

    def dx_ref(t):
        return R*w*np.cos(w*t)
    def dy_ref(t):
        return 2*R*w*np.cos(2*w*t)

    logs = []
    num_steps = 4000
    dt = 0.01

    for step in range(num_steps):
        t = step*dt
        # reference
        xr = x_ref(t)
        yr = y_ref(t)
        dxr = dx_ref(t)
        dyr = dy_ref(t)

        theta_r = np.arctan2(dyr, dxr)

        # get actual from environment
        pose = env.get_state("car1")  # e.g. { "position": [...], "heading": ..., etc. }
        x_act = pose["position"][0]
        y_act = pose["position"][1]
        theta_act = pose["heading"]  # environment provides heading directly

        # heading error
        heading_err = theta_r - theta_act
        # wrap to (-pi, pi)
        heading_err = (heading_err + np.pi) % (2*np.pi) - np.pi

        # P-control
        steering_angle_cmd = Kp_heading * heading_err
        # clamp if needed
        max_steer = 1.0
        steering_angle_cmd = np.clip(steering_angle_cmd, -max_steer, max_steer)

        env.proceed(
            vehicle_id="car1",
            speed_command=speed_cmd,
            steering_angle_command=steering_angle_cmd,
            duration=dt
        )

        logs.append([x_act, y_act])

    logs = np.array(logs)
    print("Simulation done. Closing environment.")
    env.close()

    # Reserve the final position in logs is shape (num_steps, 2)
    inputs = logs[:-1]
    outputs = logs[1:]
    N = len(inputs)
    split_idx = int(0.7*N)
    train_in, train_out = inputs[:split_idx], outputs[:split_idx]
    test_in, test_out = inputs[split_idx:], outputs[split_idx:]

    # Setup RC+UKF
    process_noise = np.eye(2)*1e-4
    measurement_noise = np.eye(2)*1e-3

    framework = RC_UKF(
        n_inputs=2,
        n_reservoir=200,
        process_noise=process_noise,
        measurement_noise=measurement_noise,
        rc_params={
            "spectral_radius": 0.9,
            "leak_rate": 1.0,
            "washout": 100,
            "reg": 1e-5,
            "random_seed": 42
        },
        ukf_params={
            "alpha": 1.0,
            "beta": 2.0,
            "kappa": 1.0
        }
    )

    print("Training reservoir on 70% of data ...")
    framework.train_reservoir(train_in, train_out)
    print("Reservoir training complete.")

    # Filter the test portion
    x0, y0 = test_in[0]
    framework.ukf.x = np.array([x0, y0])
    framework.ukf.P = np.eye(2)*1.0
    framework.reset_reservoir_state()

    meas_noise_std = 0.02
    framework.measurement_model_noise_std = 0.0

    estimated_states = []
    for i in range(len(test_in)):
        # synthetic measurement = ground truth + noise
        z = test_in[i] + np.random.normal(0, meas_noise_std, 2)
        x_est = framework.filter_step(z)
        estimated_states.append(x_est)

    estimated_states = np.array(estimated_states)
    rmse_val = compute_rmse(test_in, estimated_states)
    print(f"RC+UKF Car Figure-8 Tracking RMSE: {rmse_val:.4f}")

    # Plot
    plt.figure(figsize=(7,6))
    plt.plot(train_in[:,0], train_in[:,1], 'g-', label='Ground Truth (Train)')
    plt.plot(test_in[:,0], test_in[:,1], 'b-', label='Ground Truth (Test)')
    plt.plot(estimated_states[:,0], estimated_states[:,1], 'r--', label='RC+UKF Estimate')
    plt.title("FlightGoggles Car: Tamer Figure-8 + RC+UKF (PID-based)")
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.axis('equal')
    plt.legend()
    plt.tight_layout()
    plt.show()

if __name__ == "__main__":
    main()
```

```python
#!/usr/bin/env python
# -*- coding: utf-8 -*-

"""
flightgoggles_car_doublefreq_lissajous.py

Demonstration of a double-frequency Lissajous reference curve:
  x_ref(t) = R*sin(t),   y_ref(t) = R*sin(2*t)

In theory, a 1:2 Lissajous can form a "figure-8" if you sample exactly one period.
But in practice, with a car driving for a longer time, you'll often see multiple
vertical loops or a wavy line. Hence, it's more accurate here to call it a
"double-frequency Lissajous wave."

We use a minimal heading PID to steer the car in FlightGoggles to that path,
log the (x,y) data, then train an RC and run a UKF. The final trajectory
may not be a simple figure-8 crossing, but a repeating wave shape.
"""

import numpy as np
import matplotlib.pyplot as plt

# ------------------ FlightGoggles environment import -------------------- #
try:
    from flightgoggles.env import flightgoggles_env
except ImportError:
    raise ImportError("Could not import flightgoggles_env. Ensure the FG Python API is installed.")


def compute_rmse(true_vals, est_vals):
    return np.sqrt(np.mean((true_vals - est_vals)**2))

###############################################################################
# Minimal Classes for ESN + UKF (unchanged from your code)
###############################################################################
class ReservoirComputer:
    def __init__(
        self,
        n_inputs,
        n_reservoir=200,
        spectral_radius=0.9,
        sparsity=0.2,
        reg=1e-5,
        leak_rate=1.0,
        washout=100,
        use_bias=True,
        input_scale=0.1,
        noise_std=0.0,
        random_seed=None
    ):
        if random_seed is not None:
            np.random.seed(random_seed)

        self.n_inputs = n_inputs
        self.n_reservoir = n_reservoir
        self.spectral_radius = spectral_radius
        self.sparsity = sparsity
        self.reg = reg
        self.leak_rate = leak_rate
        self.washout = washout
        self.use_bias = use_bias
        self.input_scale = input_scale
        self.noise_std = noise_std

        # Initialize weights
        self.W_in = self._initialize_input_weights()
        self.W = self._initialize_reservoir()
        self.W_out = None

    def _initialize_input_weights(self):
        return self.input_scale * (2.0*np.random.rand(self.n_reservoir, self.n_inputs) - 1.0)

    def _initialize_reservoir(self):
        W = np.random.rand(self.n_reservoir, self.n_reservoir) - 0.5
        mask = np.random.rand(self.n_reservoir, self.n_reservoir)
        W[mask > self.sparsity] = 0.0
        eigvals = np.linalg.eigvals(W)
        max_eig = np.max(np.abs(eigvals))
        if max_eig != 0.0:
            W *= (self.spectral_radius / max_eig)
        return W

    def add_noise(self, x):
        if self.noise_std > 0:
            return x + np.random.normal(0, self.noise_std, x.shape)
        return x

    def _update_state(self, x_prev, u):
        pre_activation = self.W_in.dot(u) + self.W.dot(x_prev)
        return (1 - self.leak_rate)*x_prev + self.leak_rate*np.tanh(pre_activation)

    def run_reservoir(self, input_sequence):
        T = input_sequence.shape[0]
        x = np.zeros((self.n_reservoir,1))
        all_states = np.zeros((T, self.n_reservoir))
        for t in range(T):
            u = self.add_noise(input_sequence[t]).reshape(-1,1)
            x = self._update_state(x, u)
            all_states[t] = x.ravel()

        if self.washout >= T:
            raise ValueError("washout >= total timesteps.")
        return all_states[self.washout:]

    def train_readout(self, states, outputs):
        if self.use_bias:
            X_aug = np.hstack([states, np.ones((states.shape[0],1))])
        else:
            X_aug = states
        ridge_mat = self.reg * np.eye(X_aug.shape[1])
        self.W_out = (outputs.T @ X_aug) @ np.linalg.inv(X_aug.T @ X_aug + ridge_mat)

    def predict(self, states):
        if self.W_out is None:
            raise ValueError("Readout not trained.")
        if self.use_bias:
            X_aug = np.hstack([states, np.ones((states.shape[0],1))])
        else:
            X_aug = states
        return X_aug @ self.W_out.T

class UnscentedKalmanFilter:
    def __init__(self, n_dim, Q, R, alpha=1.0, beta=2.0, kappa=1.0):
        self.n_dim = n_dim
        self.Q = Q
        self.R = R
        self.alpha = alpha
        self.beta = beta
        self.kappa = kappa
        self.lambda_ = alpha**2*(n_dim + kappa) - n_dim

        num_sigma = 2*n_dim + 1
        self.W_m = np.full(num_sigma, 1.0/(2*(n_dim+self.lambda_)))
        self.W_c = np.full(num_sigma, 1.0/(2*(n_dim+self.lambda_)))
        self.W_m[0] = self.lambda_/(n_dim+self.lambda_)
        self.W_c[0] = self.lambda_/(n_dim+self.lambda_) + (1 - alpha**2 + beta)

        self.x = np.zeros(n_dim)
        self.P = np.eye(n_dim)

    def generate_sigma_points(self):
        n = self.n_dim
        sigma_pts = np.zeros((2*n+1, n))
        sigma_pts[0] = self.x
        U = np.linalg.cholesky((n + self.lambda_)*self.P)
        for i in range(n):
            sigma_pts[i+1]    = self.x + U[i]
            sigma_pts[n+i+1] = self.x - U[i]
        return sigma_pts

    def predict(self, process_model):
        sigma_pts = self.generate_sigma_points()
        sigma_pred = np.array([process_model(sp) for sp in sigma_pts])
        self.x = np.sum(self.W_m[:,None]*sigma_pred, axis=0)
        self.P = self.Q.copy()
        for i in range(2*self.n_dim+1):
            diff = sigma_pred[i] - self.x
            self.P += self.W_c[i]*np.outer(diff,diff)
        return sigma_pred

    def update(self, sigma_pred, z, measurement_model):
        Zsigma = np.array([measurement_model(sp) for sp in sigma_pred])
        z_mean = np.sum(self.W_m[:,None]*Zsigma, axis=0)

        P_zz = self.R.copy()
        for i in range(2*self.n_dim+1):
            diff_z = Zsigma[i] - z_mean
            P_zz += self.W_c[i]*np.outer(diff_z,diff_z)

        P_xz = np.zeros((self.n_dim, z_mean.shape[0]))
        for i in range(2*self.n_dim+1):
            diff_x = sigma_pred[i] - self.x
            diff_z = Zsigma[i] - z_mean
            P_xz += self.W_c[i]*np.outer(diff_x,diff_z)

        K = P_xz @ np.linalg.inv(P_zz)
        self.x += K.dot(z - z_mean)
        self.P -= K.dot(P_zz).dot(K.T)

    def get_state(self):
        return self.x

class RC_UKF:
    def __init__(self,
                 n_inputs,
                 n_reservoir,
                 process_noise,
                 measurement_noise,
                 rc_params=None,
                 ukf_params=None):
        if rc_params is None:
            rc_params = {}
        if ukf_params is None:
            ukf_params = {}

        self.rc = ReservoirComputer(n_inputs=n_inputs,
                                    n_reservoir=n_reservoir,
                                    **rc_params)
        self.ukf = UnscentedKalmanFilter(n_dim=n_inputs,
                                         Q=process_noise,
                                         R=measurement_noise,
                                         **ukf_params)
        self.reservoir_state = np.zeros((n_reservoir,1))
        self.measurement_model_noise_std = 0.0

    def train_reservoir(self, inputs, outputs):
        states = self.rc.run_reservoir(inputs)
        valid_outputs = outputs[self.rc.washout:]
        self.rc.train_readout(states, valid_outputs)

    def reset_reservoir_state(self):
        self.reservoir_state = np.zeros((self.rc.n_reservoir,1))

    def process_model(self, x):
        x_col = x.reshape(-1,1)
        self.reservoir_state = self.rc._update_state(self.reservoir_state, x_col)
        W_out = self.rc.W_out
        if self.rc.use_bias:
            x_next = (W_out[:, :self.rc.n_reservoir].dot(self.reservoir_state).ravel()
                      + W_out[:,-1])
        else:
            x_next = W_out[:, :self.rc.n_reservoir].dot(self.reservoir_state).ravel()
        return x_next

    def measurement_model(self, x):
        return x + np.random.normal(0.0, self.measurement_model_noise_std, x.shape)

    def filter_step(self, measurement):
        sigma_pred = self.ukf.predict(self.process_model)
        self.ukf.update(sigma_pred, measurement, self.measurement_model)
        return self.ukf.get_state()

###############################################################################
# Main function implementing the Lissajous figure-8 with heading PID
###############################################################################
def main():
    print("Starting FlightGoggles environment...")
    env = flightgoggles_env(
        cfg_dir="../config",                     
        cfg_fgclient="FlightGogglesClient_demo_car.yaml",
        cfg_car="carDynamicsSim.yaml"
    )

    # Double-frequency Lissajous reference:
    # x_ref(t)= R sin( w t ),  y_ref(t)= R sin( 2 w t )
    R = 1.0
    w = 1.0
    def x_ref(t):
        return R*np.sin(w*t)
    def y_ref(t):
        return R*np.sin(2*w*t)

    # derivatives for heading
    def dx_ref(t):
        return R*w*np.cos(w*t)
    def dy_ref(t):
        return 2*R*w*np.cos(2*w*t)

    # speed, heading PID
    speed_cmd = 0.7
    Kp_heading = 2.0
    max_steer = 1.0

    logs = []
    num_steps = 4000
    dt = 0.01

    for step in range(num_steps):
        t = step*dt
        xr = x_ref(t)
        yr = y_ref(t)
        dxr = dx_ref(t)
        dyr = dy_ref(t)

        theta_r = np.arctan2(dyr, dxr)

        # Retrieve actual state from environment
        pose = env.get_state("car1")
        x_act = pose["position"][0]
        y_act = pose["position"][1]
        theta_act = pose["heading"]  # environment stores heading directly

        # heading error
        heading_err = theta_r - theta_act
        heading_err = (heading_err + np.pi) % (2*np.pi) - np.pi

        steering_angle_cmd = Kp_heading*heading_err
        steering_angle_cmd = np.clip(steering_angle_cmd, -max_steer, max_steer)

        # proceed the environment
        env.proceed(
            vehicle_id="car1",
            speed_command=speed_cmd,
            steering_angle_command=steering_angle_cmd,
            duration=dt
        )

        logs.append([x_act, y_act])

    logs = np.array(logs)
    print("Simulation done. Closing environment.")
    env.close()

    # Now train the ESN + do a UKF on test portion
    inputs = logs[:-1]
    outputs = logs[1:]
    N = len(inputs)
    split_idx = int(0.7*N)
    train_in, train_out = inputs[:split_idx], outputs[:split_idx]
    test_in, test_out   = inputs[split_idx:], outputs[split_idx:]

    process_noise = np.eye(2)*1e-4
    measurement_noise = np.eye(2)*1e-3

    framework = RC_UKF(
        n_inputs=2,
        n_reservoir=200,
        process_noise=process_noise,
        measurement_noise=measurement_noise,
        rc_params={
            "spectral_radius": 0.9,
            "leak_rate": 1.0,
            "washout": 100,
            "reg": 1e-5,
            "random_seed": 42
        },
        ukf_params={
            "alpha": 1.0,
            "beta": 2.0,
            "kappa": 1.0
        }
    )

    print("Training reservoir on 70% of data...")
    framework.train_reservoir(train_in, train_out)
    print("Reservoir training complete.")

    # Filter the test data
    x0, y0 = test_in[0]
    framework.ukf.x = np.array([x0, y0])
    framework.ukf.P = np.eye(2)*1.0
    framework.reset_reservoir_state()

    meas_noise_std = 0.02
    framework.measurement_model_noise_std = 0.0

    estimated_states = []
    for i in range(len(test_in)):
        measurement = test_in[i] + np.random.normal(0, meas_noise_std, 2)
        x_est = framework.filter_step(measurement)
        estimated_states.append(x_est)
    estimated_states = np.array(estimated_states)

    rmse_val = compute_rmse(test_in, estimated_states)
    print(f"RC+UKF Car Double-Frequency Lissajous Wave Tracking RMSE: {rmse_val:.4f}")

    # Plot
    plt.figure(figsize=(7,6))
    plt.plot(train_in[:,0], train_in[:,1], 'g-', label='Ground Truth (Train)')
    plt.plot(test_in[:,0], test_in[:,1], 'b-', label='Ground Truth (Test)')
    plt.plot(estimated_states[:,0], estimated_states[:,1], 'r--', label='RC+UKF Estimate')
    plt.title("Double-Frequency Lissajous Wave in FlightGoggles + RC+UKF (PID-based)")
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.axis('equal')
    plt.legend()
    plt.tight_layout()
    plt.show()

if __name__ == "__main__":
    main()
```

```python
import numpy as np

def generate_figure8_trajectory(num_steps=1000, dt=0.01, A=1.0, B=1.0, w=1.0):
    """
    Generate a parametric figure-8 (Lissajous) trajectory in 2D.

    x(t) = A * sin(w * t)
    y(t) = B * sin(2 * w * t)

    Args:
        num_steps (int): Number of discrete time steps to generate.
        dt (float): Time step between consecutive samples.
        A (float): Amplitude for x direction.
        B (float): Amplitude for y direction.
        w (float): Base frequency for the sine.

    Returns:
        ts (np.ndarray): Time array of shape (num_steps,).
        x (np.ndarray): X-values of shape (num_steps,).
        y (np.ndarray): Y-values of shape (num_steps,).
    """
    ts = np.arange(num_steps) * dt
    x = A * np.sin(w * ts)
    y = B * np.sin(2.0 * w * ts)
    return ts, x, y

# Example usage and plotting (for visualization)
import matplotlib.pyplot as plt

num_steps = 1000
dt = 0.01
A = 1.0
B = 1.0
w = 1.0

ts, x, y = generate_figure8_trajectory(num_steps, dt, A, B, w)

plt.figure(figsize=(6, 6))
plt.plot(x, y)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Figure-8 Trajectory")
plt.grid(True)
plt.axis('equal') # Ensure the aspect ratio is equal for a correct figure 8.
plt.show()

#Example with more time to ensure a full figure 8 is drawn.
num_steps = 2000
dt = 0.01

ts, x, y = generate_figure8_trajectory(num_steps, dt, A, B, w)

plt.figure(figsize=(6, 6))
plt.plot(x, y)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Figure-8 Trajectory")
plt.grid(True)
plt.axis('equal')
plt.show()
```
